## Argumentos y Referencias Compartidas

```python
def f(a):	    	
	a = 99			
	
b = 88		
f(b)				
print("b = ", b)
```

Aquí la variable a está guardada en memoria local de la función f. Con lo cual al llamar fuera de la función a b pese a que parezca que tienen el mismo valor b != de a

```python
def changer(a, b):	    
		a = 2			
		b[0] = 'spam' 		
					
x = 1
l = [1, 2]
changer(x, l)           
print("l = ", l)        
assert x == 1, "x es 1"
```

l al ser unba lista y ser un valor mutable cuando modificas en b el valor si te permite cambiar su valor. Pero x al ser un valor inmutable aún cambiando el valor de a, x no se ve afectada.

```python
l = [1, 2]
changer(x, l[:]) 	

print("l = ", l) 
```

Aquí en la función changer pasamos una copia de los valores de l. Con lo cual la variable l nunca cambiará de valor.

```python
l = [1, 2]
def changer(a, b):
    b = b[:]		
    a = 2
    b[0] = 'spam'	

changer(x, l)
assert l == [1, 2], "l es una invariante"
print("l = ", l) 
```

Aquí b se define como una invariante, con lo cual al cambiar b[0] = "spam" no cambia su valor

```python
l = [1, 2]
def changer(a, b):	   
		a = 2			
		b[0] = 'spam'   
changer(x, tuple(l))	

print("l = ", l)
assert l == [1, 2], "l es una invariante"
```

Aquí b es una tupla (inmutable), por lo que b[0] = 'spam' lanza un error y l nunca se modifica, manteniéndose como invariante.

## Keyword and Default Examples

```python
def f(a, b, c): 
    print(a, b, c)

f(1, 2, 3)          
```

Se llama a la función con tres argumentos posicionales, que se imprimen tal cual, sin modificar nada.

```python
def f(a, b, c): 
    print(a, b, c)

f(c=3, b=2, a=1) 
```

Aquí se usan argumentos nombrados, así que el orden no importa; a, b, y c reciben sus valores correctamente.

```python
f(1, c=3, b=2)
```

Se pasa 1 como primer argumento posicional para a, y luego b y c como argumentos nombrados, que se asignan correctamente

```python
def f(a, b=2, c=3): 	
	print(a, b, c)


f(1)        
f(a=1) 
```

Aquí b y c tienen valores por defecto, así que si no se pasan, la función usa esos valores 

```python
def f(a, b=2, c=3): 	
	print(a, b, c)
    
f(1, 4)    
f(1, 4, 5)	

f(1, c=6)
```

f(1, 4) cambia b a 4, mantiene c por defecto (3).
f(1, 4, 5) cambia b a 4 y c a 5.
f(1, c=6) usa el valor por defecto para b (2) y cambia c a 6 con argumento nombrado.

```python
def func(spam, eggs, toast=0, ham=0):
	print((spam, eggs, toast, ham)) 	

func(1, 2)							
func(1, ham=1, eggs=0)				
func(spam=1, eggs=0)				
func(toast=1, eggs=2, spam=3)		
func(1, 2, 3, 4) 
```

spam y eggs son argumentos obligatorios, siempre hay que pasarles un valor.
toast y ham tienen valores por defecto (0), asi que son opcionales.
Puedes pasar argumentos por posicion o nombrados.
Si no pasas toast o ham usan el valor por defecto 0.
Puedes mezclar argumentos posicionales y nombrados, siempre que los posicionales vayan primero.

```python
def kwonly(a, *b, c):
    print(a, b, c)

kwonly(1, 2, c=3)       

kwonly(a=1, c=3)        

kwonly(1, 2, 3)        
```

c debe pasarse siempre con nombre, sino da error; *b recoge los extras posicionales en una tupla y a es obligatorio.

```python
def kwonly(a, *, b='spam', c='ham'):
	print(a, b, c)

kwonly(1)               
kwonly(1, c=3)          
kwonly(a=1)             
kwonly(c=3, b=2, a=1)   
kwonly(1, 2)  
```

a es argumento obligatorio y posicional o nombrado.
indica que todos los argumentos despues deben ser solo por palabra clave (keyword-only).
b y c son argumentos keyword-only con valores por defecto 'spam' y 'ham'.
En llamadas sin b o c, usan sus valores por defecto.
kwonly(1, 2) da error porque 2 se pasa como posicional, pero despues de * solo se aceptan keyword arguments.

```python
def kwonly(a, *, b, c='spam'):
    print(a, b, c)

kwonly(1, c='eggs')
```

a es argumento obligatorio posicional o nombrado.
b es obligatorio y debe pasarse por palabra clave (keyword-only).
c es keyword-only con valor por defecto 'spam'.

```python
def f(*args):
    print(args)

f()                 
f(1,)               
f(1, 2, 3, 4)       
```

args recoge todos los argumentos posicionales en una tupla llamada args.
Si no se pasa nada, args es una tupla vacia ().
Si pasas uno o mas argumentos, args los contiene todos en orden.

```python

def f(**args):
    print(args)
f()                 
f(a=1, b=2)         
```

args recoge todos los argumentos nombrados (keyword arguments) y los guarda en un diccionario args.
Si no se pasa ningun argumento nombrado, el diccionario esta vacio {}.
Si pasas argumentos nombrados, el diccionario contiene pares clave-valor con esos argumentos.

```python
def f(a, *pargs, **kargs):
    print(a, pargs, kargs)
f(1, 2, 3, x=1, y=2)        
```

a: es el primer argumento posicional obligatorio (en este caso 1).
pargs: recoge todos los argumentos posicionales adicionales en una tupla ((2, 3)).
kargs: recoge todos los argumentos nombrados (keyword arguments) en un diccionario ({'x': 1, 'y': 2}).

```python
def func(a, b, c, d): print(a, b, c, d)

args = (1, 2)
args += (3, 4)
func(*args)   
```

args es una tupla con 4 elementos: (1, 2, 3, 4)
El args en la llamada a func “desempaqueta” la tupla y pasa sus valores como argumentos separados.
Es equivalente a hacer: func(1, 2, 3, 4)

```python
args = {'a': 1, 'b': 2, 'c': 3}
args['d'] = 4
func(**args)
```

args desempaqueta el diccionario y pasa cada par clave-valor como argumento nombrado.
Es como llamar a: func(a=1, b=2, c=3, d=4)

```python
func(*(1, 2), **{'d': 4, 'c': 3})      

func(1, *(2, 3), **{'d': 4})            

func(1, c=3, *(2,), **{'d': 4})         

func(1, *(2, 3), d=4)                  

func(1, *(2,), c=3, **{'d':4})      
```

(1, 2) → pasa 1 y 2 como los primeros dos argumentos posicionales (a=1, b=2)
{'d': 4, 'c': 3} → pasa c=3, d=4 como argumentos nombrados
Equivalente a: func(1, 2, c=3, d=4)
