## Objetos Inmutables

```python
a = 3
b = a               # b es 3
print("b es ", b)

a = 3
b = a
a = 'spam'          # b es 3
print("b es ", b)
```

Si A tiene un valor y ponemos que B tenga el mismo valor que A, B solo cogera el valor que tiene A en ese momento, si A cambia de valor B mantendra el valor que cogio al principio, realmente no coge un valor solamente lo copia. 

```python
a = 3
b = a
a = a + 2     
       # b es 3
print("b es ", b)
```

Esto es como el caso de arriba, B coge el valor de A pero no copia a A solo usa el valor que copia en el momento en el que mandamos que lo haga, por eso se llaman objetos inmutables.

## Objetos Mutables

```python
L = [2, 3, 4]	    
M = L			    
L[0] = 24 		    

print("L = ", L)    
print("M = ", M)    
```

Al ser una lista es un objeto mutable, y al cambiar el valor de una lista cambia en la otra, no esta copiando el valor, estan agarrados a la misma mano en el pulpo

```python
L = [2, 3, 4]
M = L[:]
L[0] = 24

print("L = ", L)    
print("M = ", M)
```

Basicamente al usar [:] convierte la lista M en un objeto inmutable entonces al cambiar L, M no se ve afectado

## Referencias compartidas e igualidad

```python
L = [1, 2, 3]
M = L

L == M 
L is M
```

el operador igualdad valora si las dos variables tienen el mismo valor, si es exactamente el mismo valor da true como en este ejemplo y el operador Is compara si las dos variables apuntan al mismo objeto. 

```python
L = [1, 2, 3]
M = [1, 2, 3]	    

print(
    "L == M",
    L == M 			
)

print(
    "L is M", 	    
    L is M
)
```

En este caso vemos que tanto L y M tienen el mismo valor, por eso el operador igualdad da True, pero el operador Is da false porque no apuntan al mismo objeto, son dos objetos diferentes
