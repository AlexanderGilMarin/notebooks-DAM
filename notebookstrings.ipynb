## Crear una string y darle un valor

```python
guapo = "Alex"
guapo
```

## Para ver la longitud de un string

```python
len(guapo)
```

### Si quieres ver que se encuentra en una posicion especifia utilizaremos [] y dentro de los corchetes pondremos la posicion que queremos identificar, la 0 es la primera letra de la palabra

```python
guapo[0]
```

### La segunda letra seria el 1

```python
guapo[1]
```

```python
#si te sales de los limites de la palabra te da este error
guapo[4]
```

### Podemos ver la ultima letra de la palabra usando [-1]

```python
guapo[-1]
```

### La penultima letra seria usando [-2]

```python
guapo[-2]
```

## Backus-Naur
Backus-Naur es una notación formal usada para describir la sintaxis de lenguajes de programación o gramáticas formales.
El negative indexing en Python se refiere a la capacidad de acceder a elementos de una lista o cadena desde el final, usando índices negativo

```python
guapo[len(guapo) - 1]
```

### Puedes ver mas de una lentra usando estos comando

```python
guapo[1:3] #no encuentra la letra en la 3 posicion, solo la 1 y la 2 
```

```python
guapo[1:] #para ver desde el que esta en la posicion 1 hasta el final
```

```python
guapo #si os fijais no cambio 
```

```python
guapo[0:3] #utilizamos desde la primera letra, pero no cogeria la de la posicion 3, osea serian 0,1,2
```

```python
guapo[:3] #esto seria igual que la de antes.
```
